```
import keras
import pandas as pd
import numpy as np
```

```
df = pd.read_csv('train.csv')
df.head()
```

```
# X = df.drop(columns = ['SalePrice'])
X = df[['MSSubClass','LotFrontage', 'LotArea', 'YearRemodAdd', 'MasVnrArea', 'BsmtFinSF1', 'BsmtUnfSF', 'TotalBsmtSF' ]]
y = df[['SalePrice']]
```

```
model = keras.models.Sequential()
model.add(keras.layers.Dense(8, activation = 'softmax', input_shape = (8,)))
model.add(keras.layers.Dense(8, activation = 'relu'))
model.add(keras.layers.Dense(1))

model.compile(optimizer = 'adam', loss = 'mean_squared_error')
```

```
model.fit(X, y, epochs=70)
```

```
test_data = np.array([60, 65.0, 8450, 2003, 196.0, 706, 150, 856])
model.predict(test_data.reshape(1,8), batch_size = 1)
```

```
model.save('terribleModel.h5')
```
